Using plot() and scatter()

```python
plt.xlim(11,22)
plt.ylim(11**2, 22**2)
plt.plot(x,y,color='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph Title")
plt.show()
```

```python
plt.xlim(11,22)
plt.ylim(11**2, 22**2)
plt.scatter(x,y,color='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph Title")
plt.show()
```

```python
plt.plot(x,y,color='green',label='x vs y')
plt.plot(y,x,color='orange',label='y vs x')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph Title")
plt.legend(loc='upper right') #upper right/upper left, default: upper right
plt.show()
```

Using subplot()- the area is divided as 2x2 and 1st pos is given for 1st graph and 4 th position is given for 2nd graph

```python
#plt.subplot(rows,columns, division number)
plt.subplot(2,2,1)
plt.scatter(x,y,color='green')
plt.subplot(2,2,4)
plt.plot(x,y,color='purple')
```

```python
#plt.subplot(rows,columns, division number)
plt.subplot(3,3,1)
plt.plot(x,y,color='orange')
plt.subplot(3,3,8)
plt.scatter(y,x,color='pink')
plt.subplot(3,3,6)
plt.scatter(y,x,color='green')
```

Pie Graph - categorical and numerical value required

```python
courses=['AI/ML','IOT','BCT','Python']
students=[200, 175, 180, 120]
```

```python
plt.pie(students,
        labels=courses,
        autopct='%0.4f%%',#set the precision of percentage 
        colors=['purple','pink','yellow','magenta']) 
plt.show()
```

Create figure of own size using plt.figure()

```python
fig=plt.figure(figsize=(10,10))
#create the axes, the quadrants
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.plot(x,y,color='yellow')
ax2.scatter(x,y,color='purple')
ax3.bar(x,y,color='magenta')
ax4.pie(students,labels=courses,
        autopct='%0.4f%%',#set the precision of percentage 
        colors=['purple','pink','yellow','magenta'])
plt.show()
```

Using Seaborn Library- using datasets to make plots and different graphs

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips') #data of customers that tipped their waiters in a particular restaurant
tips #already available in seaborn
# some other datasets= iris
```

```python
sns.scatterplot(x='total_bill',
                y='tip',
                data=tips)
```

Inference from above SCATTER analysis:
    Data is linearly distributed, as total bill increases, tip also increases
    Higher bill amount customer tends to give higher tips

```python
sns.lineplot(x='size',y='tip', data=tips)
```

Analysis: size is directly proportional to tips

```python
sns.lineplot(x='size',y='tip', ci=0, data=tips)
```

```python
sns.barplot(x='time',y='tip',data=tips)
#inference: higher tips given during dinner time
```

```python
sns.barplot(x='day',y='tip',data=tips)
#inference: higher tips given during weekends(Saturday and Sunday)
```

```python
sns.distplot(tips['tip'])
```

```python
sns.boxplot(x='size',y='tip',data=tips)
```

```python
tips.describe()
```

```python
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
sns.catplot(x='smoker',y='tip',hue='sex',data=tips)
#hue allows two categories for any plot (except line and scatter)
```

```python
sns.catplot(x='day',y='tip',hue='sex',kind='bar',data=tips)
```

```python
sns.pairplot(tips)#only numerical values
```

Correlation- plotted using heat map

```python
tips.corr() #only numerical values
#shows how important each input is to predict each output
#>=0.5 direct proportionality (highly positively correlated)
#<0.5 not correlated, drop 
#>-0.5 not correlated, drop 
#<-0.5 indirectly proportional (highly negatively correlated)
```

```python
sns.heatmap(tips.corr(),annot=True)
#dark color, most correlation
#annot- annotation, gives values in the boxes
```

```python
sns.heatmap(tips.corr(),annot=True)
#dark color, most correlation
#annot- annotation, gives values in the boxes
```
