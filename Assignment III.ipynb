```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris') 
iris 
```

```python
sns.scatterplot(x='sepal_length',
                y='species',
                data=iris)
```

```python
sns.scatterplot(x='sepal_width',
                y='species',
                data=iris)
```

```python
sns.scatterplot(x='petal_width',
                y='species',
                data=iris)
```

```python
sns.scatterplot(x='petal_length',
                y='species',
                data=iris)
```

Scatter Plot Inferences:
    1. Petal length and width is least in Setosa, followed by Versicolor. Virginica has the longest and widest petals.
    2. Sepal length is least in Setosa, followed by Versicolor. Virginica has the longest sepals.
    3. Sepal width is least in Versicolor, followed by Virginica. Setosa has the widest sepals.

```python
sns.lineplot(x='species',y='petal_length',ci=0, data=iris)
```

```python
sns.lineplot(x='species',y='sepal_width',ci=0, data=iris)
```

Line Plot Inference: 
    Virginica has the longest petals, preceded by Versicolor then Setosa
    Setosa has the widest sepals, followed by Virginica. Versicolor has the least wide sepals.

```python
sns.barplot(x='species',y='petal_length',data=iris)
```

```python
sns.barplot(x='species',y='sepal_width',data=iris)
```

Bar Plot Inferences:
    1. Setosa has widest sepals but shortest petals.
    2. Virginica has longest petals.
    3. Versicolor has least wide sepals but average length petals.

```python
sns.distplot(iris['petal_length'])
```

Distribution Plot Inferences:
    1. Maximum number of flowers have petals between 1-2 and 3-6
    

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```

Box Plot Inferences:
    1. Petal length of Setosa lies mostly between 1 and 2 units
    2. Petal length of Versicolor lies mostly between 4 to 5 units
    3. Viriginica has the maximum petal length greatee than 5 units

```python
iris.corr()
```

Correlation Inferences:

    1. >=0.5 direct proportionality (highly positively correlated)
    2. <0.5 not correlated, drop 
    3. >-0.5 not correlated, drop 
    4. <-0.5 indirectly proportional (highly negatively correlated)



```python
sns.heatmap(iris.corr(),annot=True)
```

Heat Map Inferences:
    1. Dark color, most negative correlation 
        (sepal length and sepal width, sepal width and petal length and width, petal length and sepal width, petal width and sepal width)
    2. Light color, most positive correlation 
        (petal length and sepal length, petal length and petal width, petal width and petal length)
